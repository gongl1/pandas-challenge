## Age Demographics

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results

```python
grouped_by_sn = purchase_data_df.groupby(['SN'])
grouped_by_sn.count()
grouped_by_sn['Age'].count()
grouped_by_sn_age = pd.DataFrame(grouped_by_sn['Age'].value_counts())
grouped_by_sn_age
grouped_by_sn_age = grouped_by_sn_age.rename(columns={"Age": "Repeated Age per SN"})
grouped_by_sn_age_resetindex = grouped_by_sn_age.reset_index()
grouped_by_sn_age_resetindex
# grouped_by_sn_age_resetindex.loc[:, ["SN", "Age"]], so grouped_by_sn_age_resetindex is a DataFrame


age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

grouped_by_sn_age_resetindex["Age Group"] = pd.cut(grouped_by_sn_age_resetindex["Age"],age_bins, labels=group_names)
grouped_by_sn_age_resetindex
# grouped_by_sn_age_resetindex.loc[:, ["SN", "Age Group"]], so grouped_by_sn_age_resetindex is a DataFrame


grouped_by_sn_age_resetindex_groupbyagegroup = grouped_by_sn_age_resetindex.loc[:, ["SN", "Age", "Age Group"]]
grouped_by_sn_age_resetindex_groupbyagegroup
grouped_by_sn_age_resetindex_groupbyagegroup.loc[:, ["SN"]] # so grouped_by_sn_age_resetindex_groupbyagegroup is a DataFrame
grouped_by_sn_age_resetindex_groupbyagegroup


grouped_by_sn_age_resetindex_groupbyagegroup = grouped_by_sn_age_resetindex_groupbyagegroup.groupby(["Age Group"])
grouped_by_sn_age_resetindex_groupbyagegroup['SN'].count()
grouped_by_sn_age_resetindex_groupbyagegroup_percent = grouped_by_sn_age_resetindex_groupbyagegroup['SN'].count() / players
grouped_by_sn_age_resetindex_groupbyagegroup_percent

age_demographics = pd.DataFrame({"Total count": grouped_by_sn_age_resetindex_groupbyagegroup['SN'].count(), "Percentage of Players": grouped_by_sn_age_resetindex_groupbyagegroup_percent})
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].astype(float).map("{:.2%}".format)
age_demographics

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# group by Item ID and Item Name
grouped_by_item = purchase_data_df.groupby(['Item ID', 'Item Name'])
grouped_by_item.count()


# after grouping, under column Purchase ID, use .count() to get instance counts
grouped_by_item_purchasecount = grouped_by_item['Purchase ID'].count()
grouped_by_item_purchasecount

# after grouping, under column Price, use .mean() to get average item price
grouped_by_item_price_average = grouped_by_item['Price'].mean()
grouped_by_item_price_average

# after grouping, under column Price, use .sum() to get total price for each item
grouped_by_item_price_sum = grouped_by_item['Price'].sum()
grouped_by_item_price_sum

# create DataFrame using series above
most_popular_items_df = pd.DataFrame({
    "Purchase Count": grouped_by_item_purchasecount,
    "Item Price": grouped_by_item_price_average, 
    "Total Purchase Value": grouped_by_item_price_sum
})


most_popular_items_df_sortbypurchasecount = most_popular_items_df.sort_values("Purchase Count", ascending=False)

most_popular_items_df_sortbypurchasecount['Total Purchase Value'] = most_popular_items_df_sortbypurchasecount['Total Purchase Value'].map('${:,.2f}'.format)
most_popular_items_df_sortbypurchasecount['Item Price'] = most_popular_items_df_sortbypurchasecount['Item Price'].map('${:,.2f}'.format)

most_popular_items_df_sortbypurchasecount

```
